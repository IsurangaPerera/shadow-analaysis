```python
import base64

import requests
from IPython.display import display, Image
```

```python
# Make the API call through AWS API Gateway to get visualization data
url = ' https://dxevxkpnod.execute-api.us-east-2.amazonaws.com/prod/shadow-analysis'
response = requests.get(url)
data = response.json()
print(data["timestamp"])
```

##### Display Heatmap

```python
# Extract the heatmap from the JSON response
base64_image = data['heatmap']
# Decode the base64 image
image_bytes = base64.b64decode(base64_image)
# Display the image in Jupyter notebook
image = Image(data=image_bytes)
display(image)
```

#### Display Surface Plot

```python
# Extract the heatmap from the JSON response
base64_image = data['surface_plot']
# Decode the base64 image
image_bytes = base64.b64decode(base64_image)
# Display the image in Jupyter notebook
image = Image(data=image_bytes)
display(image)
```
